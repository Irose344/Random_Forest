```python
'''Problem Statement : You work in XYZ Company as a Python Data Scientist. The company officials
have collected some data on diabetes based on years of experience and wish for
you to create a model from it.
Dataset: diabetes.csv
'''
```

```python
#1. Load the dataset using pandas

import pandas as pd
df=pd.read_csv("diabetes.csv") #load the csv file to dataframe df
```

```python
#analyse the dataframe
df.info()
```

```python
#2. Extract data from outcome column is a variable named Y
#3. Extract data from every column except outcome column in a variable named X

Y=df['Outcome'] #target column assigned to Y
X=df.drop(columns=["Outcome"],axis=1) #all the features are assigned to X [dataframe]
```

```python
#4. Divide the dataset into two parts for training and testing in 70% and 30% proportion

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
#data is split to test and train data
#test_size=0.3 means 30 % of data -test data, rest of the 70% data-train data
#x_train,y_train---features and target value in training data
#x_test,y_test---features and target value in testing data
```

```python
#5. Create and train Decision Tree Model on training set

from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(criterion="entropy",max_depth=5,min_samples_split=5,n_estimators=50,min_samples_leaf=3)
#criterion="entropy" - entropy is used for splitting each node
#max_depth=maximum depth of the tree must be 5
#min_sample_split=minimum 5 datapoints should be present at a node for splitting
#n_estimators=50: 50 trees are created as part of model creation
#min_samples_leaf=3 : minimum 3 datapoints must be there in leaf node

model.fit(x_train,y_train)
#training model with training set
```

```python
#6. Make predictions based on the testing set using the trained model

y_pred=model.predict(x_test)
#predicted y value for testing set
```

```python
#7. Check the performance by calculating the confusion matrix and accuracy score of the model

from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
